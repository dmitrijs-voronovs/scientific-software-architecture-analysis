```python

from constants.foldernames import FolderNames
from pathlib import Path

import pandas as pd

files = Path(f"metadata/keywords/{FolderNames.ARCHITECTURE_VERIFICATION_DIR}").glob("*.csv")
df = []
for file in files:
    try:
        file_df = pd.read_csv(file)
        df.append(file_df)
    except:
        print(f"unable to read file {file}")

df = pd.concat(df)
df.source = df.source.apply(lambda x: x.split('.')[1])
# df = df.dropna(subset=["false_positive"])
df['project'] = df.apply(lambda row: f"{row['author']}/{row['repo']}/{row['version']}", axis=1)
df.head()
```

```python
df.groupby(["related_to_architecture"]).value_counts(["project"])
```

```python
arch_df = df[df["related_to_architecture"] == True].sort_values(["project", "source"])
arch_df
```

```python
arch_df.to_csv("metadata/results/architecture_verification_analysis_TRUE.csv")
```
